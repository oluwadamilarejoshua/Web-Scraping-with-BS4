```python
## Beautifulsoup for web scraping >>> Data to be used for ABC project

from cgitb import html
from bs4 import BeautifulSoup
import requests
import json

url = 'https://www.ulta.com/skin-care-cleansers?N=2794'

html_texts = requests.get(url).content
soup = BeautifulSoup(html_texts, 'lxml')
cleansers = soup.find_all('div', class_ = 'productQvContainer')
cleanser_data = []

print(cleansers)
```

```python
for cleanser in cleansers:
    category = "Skincare"
    sub_category = "Cleansers"
    try:
        product_name = cleanser.find('h4', class_ = 'prod-title').text.strip()
        product_price = cleanser.find('span', class_ = 'regPrice').text.strip()
        product_description = cleanser.find('p', class_ = 'prod-desc').text.strip()
        product_rating = cleanser.find('label', class_ = 'sr-only').text.split()[0]
        image_url = cleanser.find('img')['src']
        number_of_ratings = cleanser.find('span', class_ = 'prodCellReview').text.strip()

        product_data = {}
        product_data['Name'] = product_name
        product_data['Price'] = product_price
        product_data['Description'] = product_description
        product_data['Category'] = category
        product_data['Sub-category'] = sub_category
        product_data['Image URL'] = image_url
        product_data['Number of Ratings'] = number_of_ratings
        product_data['Rating'] = product_rating

        cleanser_data.append(product_data)
    except AttributeError as ex:
        print('Error: ', ex)
        continue
```
